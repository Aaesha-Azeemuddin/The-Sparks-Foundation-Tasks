### Task 3: Prediction using Decision Tree Algorithm

In this supervised learning task, we'll create and train a decision tree using the iris dataset. Then, we'll make predictions on new data and test the accuracy of our model.

```python
import warnings ## Import the warnings module to handle warning messages
warnings.filterwarnings("ignore") # Ignore all warning messages
```

### Importing libraries

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn.datasets as datasets
```

### Loading the dataset

```python
# Loading the iris dataset
iris=datasets. load_iris()
print ("Dataset loaded successfully")
X=iris.data[:, :]
y=iris. target
```

```python
#Input data

data=pd. DataFrame(iris['data'],columns=["Petal length", "Petal Width", "Sepal Length", "Sepal Width"])
data['Species']=iris ['target']
data['Species']=data[ 'Species'].apply(lambda x: iris ['target_names'] [x] )
data.head ()
```

```python
data.describe()
```

```python
data.info()
```

```python
# Scatter plot of data based on Sepal Length and Width features
sns. FacetGrid(data, hue='Species').map(plt.scatter, 'Sepal Length', 'Sepal Width') .add_legend ()
plt.show()
# Scatter plot of data based on Petal Length and Width features
sns. FacetGrid(data, hue='Species').map(plt.scatter, 'Petal length', 'Petal Width'). add_legend ()
plt.show()
```

# Creating model

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split

# Model Training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train,y_train)
print ("Training Complete.")
y_pred = tree_classifier.predict(X_test)
```

```python
# Comparing actual and predicted flower classification
df = pd. DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

### Measuring Accuracy

```python
# Model Accuracy
import sklearn.metrics as sm
print ("Accuracy:", sm. accuracy_score(y_test, y_pred))
```
